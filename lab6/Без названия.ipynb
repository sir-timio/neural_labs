```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
def f1(t):
    return np.cos(2*np.pi*t) * np.log(t) + t + np.random.uniform(-t/10, t/10)
def f2(t):
    return np.cos(2*np.pi*t) * np.sin(np.pi/2*t) + t + np.random.uniform(-t/10, t/10)

t1 = np.arange(1.0, 10.0, 0.1)
t2 = np.arange(1.0, 10.0, 0.02)

plt.figure(figsize=(15, 15))
plt.subplot(311)
plt.plot(t2, f1(t2), 'k')

plt.subplot(312)
plt.plot(t2, f2(t2), 'r')
plt.show()

# plt.subplot(213)
# plt.plot(t2, f3(t2), 'r--')
# plt.show()
# plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')
```
